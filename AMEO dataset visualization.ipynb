```python
import pandas as pd
import numpy as np

# Convert 'DOB' and 'DOJ' columns to datetime
df["DOB"] = pd.to_datetime(df["DOB"])
df["DOJ"] = pd.to_datetime(df["DOJ"], format='%Y-%m-%d')  # Assuming the date format is 'YYYY-MM-DD'


# Replace 'present' with current date in 'DOL' column
df.loc[df["DOL"] == "present", "DOL"] = pd.to_datetime("today")

# Convert 'DOL' column to datetime
df["DOL"] = pd.to_datetime(df["DOL"])

# Calculate tenure
df["tenure"] = df["DOL"].dt.year - df["DOJ"].dt.year
```

```python
def cdf(data):
    x = np.sort(data)
    y = np.arange(1, len(x)+1)/len(x)
    return x,y
```

```python
def univariate_num(col):
    #print basic statistics
    print(f"==={col} Statistics ===")
    print(f"Median: {df[col].median()}")
    print(f"Max: {df[col].max()}")
    print(f"Min: {df[col].min()}")
    
    #box plot
    plt.figure(figsize=(14,10))
    plt.subplot(3,2,1)
    sns.boxplot(df[col])
    plt.title(f"Box plot of {col}")
    
    #histogram and KDE plot
    plt.subplot(3,2,2)
    plt.hist(df[col], color='blue', alpha=0.7, density=True)
    sns.kdeplot(df[col], color='green',label='KDE')
    plt.title(f"Histogram and KDE of {col}")
    plt.xlabel(col)
    plt.ylabel('Density')
    plt.legend()
    
    #CDF plot
    plt.subplot(3,2,3)
    x_tenure, y_tenure = cdf(df[col])
    x_sample_tenure, y_sample_tenure = cdf(np.random.normal(df[col].mean(), df[col].std(), size=len(df[col])))
    plt.plot(x_tenure, y_tenure, linestyle='-', marker='.', color='green', label=f"{col} CDF")
    plt.plot(x_sample_tenure, y_sample_tenure, linestyle='-', marker='.', color='red', label='Normal Distribution CDF')
    plt.title(f"CDF of {col}")
    plt.xlabel(col)
    plt.ylabel('CDF')
    plt.legend()
    
    # Violin plot
    plt.subplot(3, 2, 4)
    sns.violinplot(data=df, x=col, inner='quartile')
    plt.title(f"Violin Plot of {col}")
    plt.xlabel(col)
    
    plt.tight_layout()
    plt.show()
    
```
